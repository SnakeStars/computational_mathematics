##### Постановка задачи

$$
u_t = a^2 u_{xx} + Q\delta(x)h(t_0 - t)
$$

Начальные условия:

$$
u(x,0) = 0
$$

Условия на параметры:

$$
-\infty < x < \infty, \quad t > 0
$$

Кроме того, так как стержень бесконечен и теплоизолирован:

$$
|u(x,t)| \rightarrow 0, \quad |x| \rightarrow \infty
$$

##### Аналитическое решение

$$
u(x,t) = \dfrac{Q|x|}{2kS}\left(\psi\left(\dfrac{|x|}{2\sqrt{t}}\right) - \psi\left(\dfrac{|x|}{2\sqrt{t-t_0}}\right)\eta\left(t-t_0\right)\right)
$$

$$
\psi(\xi) = \dfrac{1}{\sqrt{\pi}\xi}e^{-\xi^2} - Erf\xi
$$

##### Численное решение

Рассмотрим уравнение теплопроводности с точечным источником:

$$
\frac{\partial u}{\partial t} = a^2 \frac{\partial^2 u}{\partial x^2} + Q \delta(x) h(t_0 - t),
$$

где:
- $a^2$ — коэффициент теплопроводности,  
- $\delta(x)$ — точечный источник в $x=0$,  
- $h(t_0 - t)$ — функция Хевисайда (источник активен при \(t < t_0\)).

##### 1. Аппроксимация второй производной на неравномерной сетке

Пусть сетка $x_0 < x_1 < \dots < x_n$ с неравномерными шагами

$$
h_i^- = x_i - x_{i-1}, \quad h_i^+ = x_{i+1} - x_i.
$$

Тогда вторая производная аппроксимируется как:

$$
u_{xx}(x_i) \approx \frac{2}{h_i^+ + h_i^-} \left( \frac{u_{i+1}-u_i}{h_i^+} - \frac{u_i-u_{i-1}}{h_i^-} \right).
$$

##### 2. Вид системы на некотором шаге

Схема:

$$
\frac{u_i^{n+1}-u_i^n}{\Delta t} = a^2\left(\left(\nu\right)u_{xx}^{n+1} + \left(1 - \nu\right)u_{xx}^{n}\right) + S_i^n,
$$

где дискретный источник:

$$
S_i^n =
\begin{cases}
\dfrac{Q}{\Delta x_0} \, h(t_0 - t^n), & i = i_0, \\
0, & i \ne i_0,
\end{cases}
$$

- $i_0$ — индекс узла около нуля


##### Схема Кранка -- Николсон для уравнения:

Уравнение:

$$
\frac{u_i^{n+1}-u_i^n}{\Delta t} = a^2 \left( \nu u_{xx}^{n+1} + (1-\nu) u_{xx}^{n} \right) + S_i^n
$$

Аппроксимация второй производной на неравномерной сетке:

$$
u_{xx}(x_i) \approx \frac{2}{h_i^+ + h_i^-} \left( \frac{u_{i+1}-u_i}{h_i^+} - \frac{u_i-u_{i-1}}{h_i^-} \right)
$$

##### Коэффициенты трёхдиагональной матрицы (для n+1)

$$
\begin{aligned}
A_i &= - \frac{2 a^2 \nu \Delta t}{(h_i^+ + h_i^-) h_i^-}, \\
C_i &= - \frac{2 a^2 \nu \Delta t}{(h_i^+ + h_i^-) h_i^+}, \\
B_i &= 1 + \frac{2 a^2 \nu \Delta t}{h_i^+ + h_i^-} \left( \frac{1}{h_i^-} + \frac{1}{h_i^+} \right)
\end{aligned}
$$

Правая часть (слой n):

$$
R_i = u_i^n + a^2 (1-\nu) \Delta t \cdot u_{xx}^n + \Delta t \, S_i^n
$$

- Источник: 
$$
S_i^n = 
\begin{cases}
\dfrac{Q}{\Delta x_0} h(t_0 - t^n), & i = i_0 \\
0, & i \neq i_0
\end{cases}
$$

##### 4. Трёхдиагональная система

Собираем матрицу для всех внутренних узлов \(i=1,\dots,n-1\):

$$
\begin{bmatrix}
B_1 & C_1 & 0 & \dots & 0 \\
A_2 & B_2 & C_2 & \dots & 0 \\
0 & A_3 & B_3 & \dots & 0 \\
\vdots & & & \ddots & C_{n-2} \\
0 & \dots & 0 & A_{n-1} & B_{n-1} 
\end{bmatrix}
\begin{bmatrix}
u_1^{n+1} \\ u_2^{n+1} \\ u_3^{n+1} \\ \vdots \\ u_{n-1}^{n+1}
\end{bmatrix}
=
\begin{bmatrix}
R_1 \\ R_2 \\ R_3 \\ \vdots \\ R_{n-1}
\end{bmatrix}.
$$

##### 5. Пошаговое решение

1. Вычисляем коэффициенты \(A_i, B_i, C_i\) для всех узлов.  
2. Формируем правую часть \(R_i\) с учётом источника \(S_i^n\).  
3. Решаем трёхдиагональную систему (например, методом `solve_banded` в Python):
4. Переходим к следующему временному шагу.
